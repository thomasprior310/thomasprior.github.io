```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = r"C:\Users\thoma\Portfolio Projects\NBA_25_Greatest_Games.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
print(df.head())
```

```python
#Convert game_date column to datetime
df['game_date'] = pd.to_datetime(df['game_date'])
```

```python
#Create Histogram
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.hist(df['game_date'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Game Dates')
plt.xlabel('Game Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
